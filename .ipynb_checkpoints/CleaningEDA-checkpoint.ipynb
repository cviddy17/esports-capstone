```python
def compare_date_range(a, d):
    if (a > d[0]) & (a < d[1]):
        return True
    else:
        return False
```

```python
def team_name_processing(dict, end_date):
    df = pd.DataFrame(dict)
    a = np.datetime64(end_date)
    for i in range(len(df)):
        b = df['history'][i]
        b = parser(b)
        if compare_date_range(a, b):
            return BeautifulSoup(df.index[i],'lxml').text
    return 'Unaffiliated'
```
