```python
import pandas as pd
import numpy as np
```

```python
countries = pd.read_csv('countries.csv')
```

```python
ct = countries.set_index('alpha-2')
```

```python
def region(cc):
    if (cc != 'Other'):
        t = ct['region'][cc.upper()]
        return t
    else:
        return cc
```

```python
messy = pd.read_json('messy.json')
```

```python
def subregion(cc):
    if (cc != 'Other'):
        t = ct['sub-region'][cc.upper()]
        return t
    else:
        return cc
```

```python
messy['region']= messy['CountryCode'].apply(region)
```

```python
messy['sub-region']= messy['CountryCode'].apply(subregion)
```

```python
def name(cc):
    if (cc != 'Other'):
        t = ct['name'][cc.upper()]
        return t
    else:
        return cc
messy['CountryName']=messy['CountryCode'].apply(name)
messy.head()
```

```python
messy.to_json('messy_with_countries.json')
```
