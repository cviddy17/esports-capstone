```python
import requests
import pandas as pd
import numpy as np
import json
import time
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
df = pd.read_csv('starcraft_names_with_keys.csv')
```

```python
for i in np.arange(94, df.shape[0]):
    j = {}
    j['PlayerId'] = df['PlayerId'][i]
    url = 'https://www.esportsearnings.com/players/'+df['Key'][i]+'/team-history'
    urlop = requests.get(url)
    soup = BeautifulSoup(urlop.text, 'lxml')
    try:
        b = soup.select('div.format_cell.info_text_value')[1]
        dob = (b.contents[0])
        j['dob']=dob
    #     soup2 = soup.select('table')
    #     soup3 = soup2[0].select('td.detail_list_date')
    #     soup4 = soup2[0].select('td.detail_list_player')
    #     d = {}
    #     for i in range(len(soup3)):
    #         s4 = str(soup4[i].contents)
    #         d[s4] = str(soup3[i].contents)
    #     j['info']=d
        print(i, j)
        df2 = pd.DataFrame(j,index=[0])
        df2.to_csv('dob.csv',mode='a',header=False)
        time.sleep(2)
    except:
        time.sleep(2)
```
