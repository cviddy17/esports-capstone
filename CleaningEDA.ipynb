```python
def compare_date_range(a, d):
    a = np.datetime64(a)
    d[0] = np.datetime64(d[0])
    d[1] = np.datetime64(d[1])
    if (a > d[0]) & (a < d[1]):
        return True
    else:
        return False
```

```python
def team_name_processing(l, end_date):
    '''Takes in a list of dictionaries, returns team'''
    a = np.datetime64(end_date)
    for i in l:
        b = i['history']
#         b = parser(b)
        if compare_date_range(a, b):
            return i['team']
    return 'Unaffiliated'
```
